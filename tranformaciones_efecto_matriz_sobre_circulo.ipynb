## Haciendo la transformacion paso a paso por UDV

```python
U,D,V = np.linalg.svd(A)
print(U.shape,D.shape,V.shape)
```

```python
print('Primera rotacion V')
graficarMatriz(V)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
print('Escala D')
graficarMatriz(np.diag(D).dot(V))
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```

```python
print('Segunda rotacion U')
graficarMatriz(U.dot(np.diag(D).dot(V)))
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```
